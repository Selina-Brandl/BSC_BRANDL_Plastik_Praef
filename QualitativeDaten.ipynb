```r
library(plyr)
library(reshape2)
library(gridExtra)

#lade csv file als data.frame rawData
rawData = read.csv("QualitativeDaten.csv", header = TRUE, stringsAsFactors = FALSE)

#erstelle neue Spalte mit Produkttyp ("Plastik" vs "Alternativ")
rawData$Typ = "Plastik"
rawData[startsWith(rawData$Produkt, "A"), "Typ"] = "Alternativ"

#erstelle neue Spalte mit Produktkategorie
rawData$Kategorie = substr(rawData$Produkt, 2, 2)
rawData$Kategorie = as.integer(rawData$Kategorie)

#daten anzeigen
head(rawData,20)

#Zusammenfassung anzeigen
summary(rawData)
```

```r
#Datenformat normieren
data = melt(rawData, id.var = c("VP", "Geschlecht", "VB", "Produkt", "Kategorie", "Typ", "Frage"), variable.name = "Antwort", value.name = "Wert")
data = data[data$Wert != '',]
data$Antwort <- NULL

#Daten anzeigen
head(data)

#Zusammenfassung anzeigen
summary(data)
```

```r
casted = dcast(data[data$Frage=="B1_Welche Infos",], Produkt~Wert)
casted

casted2 = dcast(data[data$Frage=="B1_Welche Infos",], Typ~Wert)
casted2

pdf("B1_Welche Infos.pdf", height=4, width=15)
grid.table(casted)
dev.off()

pdf("B1_Welche Infos_Typ.pdf", height=4, width=15)
grid.table(casted2)
dev.off()
```
