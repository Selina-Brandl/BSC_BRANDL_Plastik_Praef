```r
library(reshape2)
library(ggplot2)
library(ggrepel)

#lade csv file als data.frame rawData
rawData = read.csv("KostenVergleich.csv", header = TRUE, stringsAsFactors = FALSE)

#erstelle neue Spalte mit Produkttyp ("Plastik" vs "Alternativ")
rawData$Typ = "Plastik"
rawData[startsWith(rawData$Produkt, "A"), "Typ"] = "Alternativ"

#erstelle neue Spalte mit Produktkategorie
rawData$Kategorie = substr(rawData$Produkt, 2, 2)
rawData$Kategorie = as.integer(rawData$Kategorie)

print(rawData)
```

```r
#Datenformat normieren
data = melt(rawData, id.var = c("VP", "Produkt", "Preis.Echt.Gesch√§tzt", "Typ", "Kategorie"), variable.name = "Frage", value.name = "Wert")
data$Frage = as.character(data$Frage)

#Daten anzeigen
head(data)

#usammenfassung anzeigen
summary(data)
```

```r
#Funktion f√ºr line chart f√ºr ein feature
drawAvgDataFeatureByProductLineChart = function(feature, typ){
  # Datenset selektieren
  dataFeature = data[data$Frage == feature & data$Typ == typ,]
  # Durchschnitt pro Frage berechnen
  avgDataFeatureByProduct = with(dataFeature, aggregate(Wert ~ Produkt + Preis.Echt.Gesch√§tzt + Typ + Frage, FUN = mean))
  # Line Charts f√ºr avgDataFeatureByProduct erstellen
  avgDataFeatureByProductLineChart = ggplot(avgDataFeatureByProduct, aes(x = Preis.Echt.Gesch√§tzt, y = Wert, label = Produkt, group = Produkt, color = Preis.Echt.Gesch√§tzt)) + 
    geom_line(color='black') +
    geom_point() +
    geom_label_repel()+
    scale_color_manual(values=c("#c98a00", "#2b61cc"))
  
  print(avgDataFeatureByProduct)
  print(avgDataFeatureByProductLineChart)
}
```

```r
#erstelle line chart f√ºr B2_Kosten
drawAvgDataFeatureByProductLineChart("Kosten", "Plastik")

ggsave("B2_Kosten_Plastik.png")
```

```r
#erstelle line chart f√ºr B2_Kosten
drawAvgDataFeatureByProductLineChart("Kosten", "Alternativ")

ggsave("B2_Kosten_Alternativ.png")
```
